```
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

scores = np.array([65, 70, 75, 80, 85, 90, 78, 82, 88, 92, 76, 84, 79, 73, 87])
print(scores)
```

```
filtered = scores[scores > 80]
print(filtered)
```

```
mean = np.mean(scores)
median = np.median(scores)
mode = stats.mode(scores)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode.mode)
```

```
data_range = np.max(scores) - np.min(scores)
variance = np.var(scores)
std_dev = np.std(scores)

Q1 = np.percentile(scores, 25)
Q3 = np.percentile(scores, 75)
IQR = Q3 - Q1

print("Range:", data_range)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("IQR:", IQR)
```

```
plt.hist(scores)
plt.title("Histogram of Scores")
plt.show()

plt.boxplot(scores)
plt.title("Boxplot of Scores")
plt.show()
```

```
scores_outlier = np.append(scores, 150)

print(scores_outlier)
```

```
mean2 = np.mean(scores_outlier)
median2 = np.median(scores_outlier)
variance2 = np.var(scores_outlier)
std2 = np.std(scores_outlier)

Q1_2 = np.percentile(scores_outlier, 25)
Q3_2 = np.percentile(scores_outlier, 75)
IQR2 = Q3_2 - Q1_2

print("Mean:", mean2)
print("Median:", median2)
print("Variance:", variance2)
print("Std Dev:", std2)
print("IQR:", IQR2)
```

```
import pandas as pd

df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

df2 = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [40, 50, 60]
})

add_result = df1 + df2
sub_result = df1 - df2

print("Addition:\n", add_result)
print("\nSubtraction:\n", sub_result)
```

```
import numpy as np

df = pd.DataFrame({
    "A": [4, 9, 16],
    "B": [1, 25, 36],
    "C": ["x", "y", "z"]
})

df[["A", "B"]] = df[["A", "B"]].apply(np.sqrt)
print(df)
```

```
df = pd.DataFrame({
    'A': [1, 5, 9],
    'B': [2, 6, 10]
})

range_result = df.apply(lambda x: x.max() - x.min())

print(range_result)
```

```
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], index=['b', 'c', 'd'])

print("Default addition:\n", s1 + s2)

print("\nWith fill_value=0:\n", s1.add(s2, fill_value=0))
```

```
df = pd.DataFrame({
    'A': [10, 20, 30, 40, 50]
})

mean = df['A'].mean()
std = df['A'].std()

df['z_score'] = (df['A'] - mean) / std

print(df)
print("\nMean of z:", df['z_score'].mean())
print("Std of z:", df['z_score'].std())
```

```
df = pd.DataFrame({
    'price': [100, 200, 300],
    'quantity': [1, 2, 3]
})

# Feature engineering
df['total'] = df['price'] * df['quantity']
df['discounted_price'] = df['price'] * 0.9
df['log_price'] = np.log(df['price'])

print(df)
```

```
import pandas as pd
import numpy as np

s = pd.Series([1, None, 3, np.nan, 5])

print(s)
```

```
print(s.isna())
print(s.notna())
```

```
df = pd.DataFrame({
    "A": [1, np.nan, 3],
    "B": [4, 5, np.nan],
    "C": [np.nan, np.nan, 9]
})

print(df)
```

```
print(df.dropna())
```

```
print(df.shape)
print(df.dropna().shape)
print(df.dropna(axis=1).shape)
```

```
print(df.fillna(0))
```

```
print(df.fillna(df.mean()))
```

```
s = pd.Series([1, np.nan, np.nan, 4, np.nan])

print(s.fillna(method="ffill"))
print(s.fillna(method="bfill"))
```

```
df = pd.DataFrame({
    "A": [1, np.nan, 3],
    "B": [4, 5, np.nan],
    "C": [7, np.nan, 9]
})

print(df)
```

```
print(df.dropna(thresh=2))
print(df.dropna(subset=["A"]))
```

```
df = pd.DataFrame({
    "A": [1, np.nan, 3],
    "B": [np.nan, 5, np.nan],
    "C": [7, 8, 9]
})

missing_count = df.isna().sum()
missing_percent = (df.isna().mean()) * 100

summary = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing %": missing_percent
})

print(summary)
```

```
import pandas as pd


index = [('A', 1), ('A', 2), ('B', 1), ('B', 2)]


multi_index = pd.MultiIndex.from_tuples(index)


s = pd.Series([10, 20, 30, 40], index=multi_index)

print(s)
print(s.loc['A'])
```

```
df = s.unstack()
print(df)
```

```
print(df.stack())
```

```
s.index.names = ['letter', 'number']
print(s)
```

```
print(s.index.get_level_values('letter'))
```

Advenced


```
index = [
    ('USA', 'NY'),
    ('USA', 'LA'),
    ('UK', 'London'),
    ('UK', 'Manchester')
]

multi_index = pd.MultiIndex.from_tuples(index, names=['country', 'city'])

df = pd.DataFrame({'population': [8, 4, 9, 3]}, index=multi_index)

print(df)
print(df.loc['USA'])
```

```
df_swapped = df.swaplevel()
print(df_swapped)


df_sorted = df_swapped.sort_index()
print(df_sorted)

print(df_sorted.loc['NY'])
```

```
print(df.groupby(level='country').mean())
print(df.groupby(level='country').sum())
```

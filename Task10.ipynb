```
import numpy as np
students = np.array([
    ('Kairat', 20, 98),
    ('Aidai', 21, 92),
    ('Beksultan', 19, 78),
    ('Nurbek', 22, 95),
    ('Diana', 20, 88)
], dtype=[('name', 'U10'), ('age', 'i4'), ('score', 'i4')])
```

```
avg = students['score'].mean()

filtered = students[students['score'] > avg]
print(avg)
print(filtered)
```

```
sorted_students = np.sort(filtered, order=['score', 'age'])[::-1]
sorted_students1 = np.sort(filtered, order=['age', 'score'])[::]
print(sorted_students)
print(sorted_students1)
```

```
top3 = sorted_students[:3]

print(top3[['name', 'score']])
```

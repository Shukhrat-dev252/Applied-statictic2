```
import pandas as pd
from io import StringIO

csv_data = """id,name,age,score
1,Ali,20,85
2,Aida,21,92
3,Bek,19,78
"""

df = pd.read_csv(StringIO(csv_data))

print(df)
```

```
df = df.set_index('id')
print(df)
```

```
df = df.sort_index()
print(df)
```

```
df['double_score'] = df['score'] * 2
df['passed'] = df['score'] > 80

print(df)
```

DataFrame vs Series
Series = one column (1D data)
DataFrame = table (multiple columns)
✔ Index alignment
Pandas aligns data using index labels, not position
Example: adding two columns matches rows by index automatically
✔ Why vectorization is faster
Works on whole column at once
Uses optimized C code internally
No Python loops → much faster 🚀
